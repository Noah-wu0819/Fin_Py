```python
########################################################################################
#getting data from coingecko//completely free!!!
from pycoingecko import CoinGeckoAPI
from datetime import datetime
import pandas as pd
cg = CoinGeckoAPI()
```

```python
ohlc = cg.get_coin_ohlc_by_id(
    id = "bitcoin", vs_currency = "usd", days = "14"
)
ohlc_df = pd.DataFrame(ohlc)
ohlc_df.columns = ["date", "open", "high", "low", "close"]
ohlc_df["date"] = pd.to_datetime(ohlc_df["date"], unit = "ms")
ohlc_df#we must predefine number of days
#varying frequency: 
#30 minutes for requests of 1 or 2 days
#4 hours between 3 and 30 days
#every 4 days above 30 days
```

```python
#get the top 7 trending coins
trending_coins = cg.get_search_trending()
(
    pd.DataFrame([coin["item"] for coin in trending_coins["coins"]])
    .drop(columns=["thumb","small","large"])
)
#double braces means the return of the front function is a function which needs parameters
#in the back brace
```
