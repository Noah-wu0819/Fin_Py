```python
#1.Converting prices to returns
#Benefit:1. making the time series stationary 2. normalization
#By converting prices into returns, we can better analyze and predict the price trends of 
#financial assets and their performance in different time periods
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# download data from Yahoo Finance
df = yf.download("AAPL",
                start = "2010-01-01",
                end = "2020-12-31",
                progress = False)

df = df.loc[:, ["Adj Close"]]
```

```python
#calculate the simple and log returns using the adjusted close prices:
df["simple_rtn"] = df["Adj Close"].pct_change()
df["log_rtn"] = np.log(df["Adj Close"]/df["Adj Close"].shift(1))
df.head()
#In the case of working with time series data, we need to make sure that the data is sorted by the time index
```

```python
#Adjusting the returns for inflation
'''
Inflation refers to the phenomenon where  money loses purchasing power over time and prices rise.If
we don't consider infation, we may overestimate the returns of financial assets because we are not 
accounting for the impact of rising prices. Therefore, we need to adjust the returns of financial 
assets for inflation in order to estimate their performance more accurately
'''

```

```python
df_cpi = (
    nasdaqdatalink.get(dataset="RATEINF/CPI_USA",
                      start_date = "2009-12-31",
                      end_date = "2020-12-31")
    .rename(columns={"Value":"cpi"})
)
'''
RATEINF/CPI_USA is the name of a dataset on the FRED (Federal Reserve Economic Data) website 
that contains Consumer Price Index (CPI) data for the United States. 
"RATEINF" is one of the categories for indicators in FRED, 
representing data related to inflation and inflation rates. 
CPI is a measure of the price level of a basket of goods 
and services typically consumed by households, 
and is used as an indicator of changes in the cost of living over time.
'''
df_cpi
```

```python
df = df.join(df_cpi, how = "left")
df["simple_rtn"] = df["Adj Close"].pct_change()
df["inflation_rate"] = df["cpi"].pct_change()
df["real_rtn"] = (
    (df["simple_rtn"]+1) / (df["inflation_rate"]+1) - 1
)
df.head()
```

```python
import pandas as pd
import numpy as np
```

```python
def realized_volatility(x):
    return np.sqrt(np.sum(x**2))
```

```python
df = yf.download("AAPL",
                start = "2000-01-01",
                end = "2012-12-31",
                progress = False)
df = df.loc[:,["Adj Close"]]
df["log_rtn"] = np.log(df["Adj Close"]/df["Adj Close"].shift(1))
df.drop(labels = "Adj Close", axis = 1, inplace = True)
df
```

```python
df_rv = (
    df.groupby(pd.Grouper(freq = "M"))
    .apply(realized_volatility)
    .rename(columns={"log_rtn":"rv"})
)
df_rv
```

```python
df_rv.rv = df_rv["rv"] * np.sqrt(12)
df_rv
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 1, sharex = True)
ax[0].plot(df)
ax[0].set_title("Apple's log returns(2000 - 2012)")
ax[1].plot(df_rv)
ax[1].set_title("Annualized realized volatitlity")
plt.show()
```

```python
#different ways of imputing missing data
import pandas as pd
import numpy as np
import nasdaqdatalink
nasdaqdatalink.ApiConfig.api_key = "uLwMFhygQHyTJGEJ6QM3" 
df = (
    nasdaqdatalink.get(dataset="RATEINF/CPI_USA",
                       start_date="2015-01-01", 
                       end_date="2020-12-31")
    .rename(columns={"Value": "cpi"})
)
```

```python
np.random.seed(42)
rand_indices = np.random.choice(df.index, 5, replace = False)

'''

np.random.choice(df.index, 5, replace = False)is used to randomly select 5 samples from the index of df data 
without replacement

'''
```

```python
df["cpi_missing"] = df.loc[:, "cpi"]
df.loc[rand_indices, "cpi_missing"] = np.nan
df.head()
```

```python
for method in ["bfill", "ffill"]:
    df[f"method{method}"]= (
        df[["cpi_missing"]].fillna(method = method)
    )
```

```python
df.loc[rand_indices].sort_index()
```

```python
df.loc[:"2017-01-01"]\
    .drop(columns = ["cpi_missing"])\
    .plot(title = "Different ways of filling missing values")
```

```python

df.loc[:"2017-01-01"]\
    .drop(columns = ["cpi_missing"])\
    .plot(title = "Different ways of filling missing values")
```

```python
df["method_interpolate"] = df[["cpi_missing"]].interpolate()
df.loc[rand_indices].sort_index()
df.loc[:"2017-01-01"]\
    .drop(columns=["cpi_missing"])\
    .plot(title="Different ways of filling missing values")
```
